### Система где одно из собственных значений равно 0

Дана система линейных уравнений:

\begin{array}{l} U'=-2*U+V, \\ V'=6*U-3*V.  \end{array}

Ссылка: https://cudahub.accelcomp.org/user/hpc/notebooks/cudahub/tracer/hybriddomain/tests/prosto_fazportret1.ipynb

Результат: Совпадает с действительностью, показывает траекторию ввиед отрезка конец которой приближается к прямой состояний равновесия.

### Система с циклами запуск траектории между ними

Дана система уравнений:

\begin{array}{l} U'=U*(U^2+V^2)-V, \\ V'=V*(U^2+V^2)+U.  \end{array}

Ссылка: https://cudahub.accelcomp.org/user/hpc/notebooks/cudahub/tracer/hybriddomain/tests/prosto_fazportret2.ipynb

Результат: Совпадает с действительностью, начальное вращение траектории описывает первый.


### Система с тремя состониями равновесия

Дана система уравнений:

\begin{array}{l} U'=U*V-4, \\ V'=(U-4)*(V-U).  \end{array}

Ссылка: https://cudahub.accelcomp.org/user/hpc/notebooks/cudahub/tracer/hybriddomain/tests/prosto_fazportret3.ipynb

Результат: Совпадает с действительностью, начальное вращение траектории описывает движение от неустойчивого фокуса к устойчивому, заметна разница между решением трейсера и решением библиотечного метода.


### Система описывающая хищник-жертва (первый вариант)

Дана система уравнений:

\begin{array}{l} U'=2*U-U*V, \\ V'=-2*V+U*V.  \end{array}
